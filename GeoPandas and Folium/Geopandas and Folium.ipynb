```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("CovidDIY.geojson")
gdf.head()
```

```python
gdf = gdf[['Kabupaten/Kota', 'Positif', 'Sembuh', 'Meninggal','Odp','Pdp','Terakhir Update', 'geometry']].fillna(0)
```

```python
pip install folium
```

```python
import folium
m = gdf.explore(column="Kabupaten/Kota", name="Kabupaten")
folium.LayerControl().add_to(m)
m
```

```python
m.save("DC Map.html")
```

```python
m = folium.Map(
    location=[-7.794914855530829, 110.37106715251684],
    tiles='Cartodb Positron',
    attr='Farrel Map',  # Ganti dengan teks attribusi yang sesuai
    zoom_start=11
)

# Sisanya dari kode yang sebelumnya
tooltip = folium.features.GeoJsonTooltip(
    fields=['Kabupaten/Kota', 'Positif'],
    aliases=["Wilayah", "Terkena Covid"],
    labels=True,
    sticky=False
)
folium.GeoJson(gdf, tooltip=tooltip).add_to(m)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(
    location=[-7.794914855530829, 110.37106715251684],
    tiles='Cartodb Positron',
    attr='Attribution Text Here',  # Ganti dengan teks attribusi yang sesuai
    zoom_start=11
)

cpleth = folium.Choropleth(gdf, data=gdf, key_on='feature.properties.Kabupaten/Kota',
                          columns=['Kabupaten/Kota', 'Positif'],
                          fill_color="RdYlGn",
                          legend_name="Percent Change, 2020 ", 
                          name="DIY Covid")
cpleth.geojson.add_child(tooltip)
cpleth.add_to(m)
folium.LayerControl().add_to(m)
m
```
